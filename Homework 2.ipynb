4.a Extract the latitudes and longitudes as floats

```python
df_police = pd.read_csv('Downloads/Police_Stations.csv')
```

```python
df_police['Station Latitude'] = ""
df_police['Station Longitude'] = ""
df_police_location = df_police['LOCATION'].astype(str)
location = df_police_location.str.split('(').apply(pd.Series).astype(str)
location = location[1]

latlong = location.str.split(')').apply(pd.Series).astype(str)
latlong = latlong[0]

latlong2 = latlong.str.split(',').apply(pd.Series).astype(str)
latitude = latlong2[0].astype(float)
longitude = latlong2[1].astype(float)
```
