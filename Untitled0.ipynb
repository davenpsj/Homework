2.a join census block and community mapping area

```python
df_census = pd.read_csv('Downloads/census.csv')
df_census.columns
```

```python
df_tracts = pd.read_csv('Downloads/tracts.csv')
df_tracts.columns
```

```python
df_census['CENSUS BLOCK'] = df_census['CENSUS BLOCK'].astype(str)
df_tracts['tract_id'] = df_tracts['tract_id'].astype(str)
df_tracts['tract_id'] = df_tracts['tract_id'].str[5: ]
                                  
def add_0(x):
    if len(x) == 9:
        return '0' + x
    elif len(x) == 10:
        return x
```

```python
df_census['CENSUS BLOCK'] = df_census['CENSUS BLOCK'].apply(add_0)
df_census['CENSUS BLOCK'] = df_census['CENSUS BLOCK'].str[0:6]
```

```python
df_census_aggregate = df_tracts.merge(df_census, left_on=['tract_id'], right_on=['CENSUS BLOCK'], how='inner')
```

2.b calculate the total population in each community area

```python
community_id = df_census_aggregate.groupby(' community_id')
pop_by_community = community_id['TOTAL POPULATION'].sum()
df_pop_by_community = pd.DataFrame(pop_by_community)
df_pop_by_community
```
